### Introduction and Project Description:
Analysis of neighborhoods in Scarborough ( Toronto ) based on the distribution of various facilities in and around that neighborhood to help the new Neighbors to take a better decision on choosing the best neighborhood out of Scarborough.

The main purpose of this project is to help the new neighbors to explore the various neighborhoods before actually move in to any neighborhood. 

As an example, two randomly picked neighborhoods viz. Agincourt, Cedarbrae are compared and analyzed the top 10 most common venues in each of those two neighborhoods based on the number of visits by people in each of those places. 

Normally people tend to explore the places before moving to a new country, state, city or place for their work or to start a new life and that involves so many factors including neighborhood analysis. That calls for a search algorithm that usually returns the requested features such as population rate, median house price, school ratings, crime rates, weather conditions, recreational facilities etc. It would be nice to have a hassle-free application that could spit out an extensive analysis of all these features for a neighborhood or a comparative analysis between neighborhoods with just sending out the names of the neighborhoods.
